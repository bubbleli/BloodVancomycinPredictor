```python
from datetime import datetime

datetimeFormat = '%Y-%m-%d %H:%M:%S'     
```

```python
def prev_dose(x):
  arr = time_dict[x['hadm_id']]
  result = arr[0][0]
  if x['event'][2] == 'dose':
    return x['event'][0]
  for i in arr:
    if i[0] < x['time'] and i[0] > result and i[2] == 'dose':
      result = i[0]
  return result
```

```python
def prev_dose_amount(x):
  arr = time_dict[x['hadm_id']]
  result_time = arr[0][0]
  result = 0 
  if x['event'][2] == 'dose':
    return x['event'][1]
  for i in arr:
    if i[0] <= x['time'] and i[0] >= result_time and i[2] == 'dose':
      result_time = i[0]
      result = i[1]
  return result
```

```python
def prev_creatinine_amount(x):
  arr = time_dict[x['hadm_id']]
  result_time = arr[0][0]
  result = 0 
  if x['event'][2] == 'dose':
    return x['event'][3]
  for i in arr:
    if i[0] <= x['time'] and i[0] >= result_time and i[2] == 'dose':
      result_time = i[0]
      result = i[3]
  return result
```

```python
def prev_bun_amount(x):
  arr = time_dict[x['hadm_id']]
  result_time = arr[0][0]
  result = 0 
  if x['event'][2] == 'dose':
    return x['event'][4]
  for i in arr:
    if i[0] <= x['time'] and i[0] >= result_time and i[2] == 'dose':
      result_time = i[0]
      result = i[4]
  return result
```

```python
def next_vanco(x):
  arr = time_dict[x['hadm_id']]
  result = arr[-1][0]
  for i in arr:
    if i[0] > x['time'] and i[0] < result and i[2] == 'test':
      result = i[0]
  return result
```

```python
def next_vanco_level(x):
  arr = time_dict[x['hadm_id']]
  result_time = arr[-1][0]
  result = 0
  for i in arr:
    if i[0] > x['time'] and i[0] <= result_time and i[2] == 'test':
      result_time = i[0]
      result = i[1]
  return result
```

```python
def curr_vanco_level(x):
    arr = time_dict[x['hadm_id']]
    if x['event'][2] == 'test':
        return x['event'][1]
    else:
        return 0
```
